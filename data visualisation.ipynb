```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,1,0.05)
y=x*x
z=x*x*x
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
#plt.plot() completes the graph
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
```

```python
#customising the graphs...
plt.figure(figsize=(12,5))
plt.plot(x,y,label='x v/s y')
plt.plot(x,z,label='x v/s z')
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.title("graph bn x,y and z")
plt.legend()#to show lables
plt.show()
```

```python
#scatter plot or discrete plot
plt.figure(figsize=(12,5))
plt.scatter(x,y,c='r')
plt.plot(x,y,c='g')
plt.show()
```

```python
#historgram-frequency distribution
data=[25,26,24,25,26,23,21,24,25,28,29,27,25,26]
plt.figure(figsize=(10,4))
plt.hist(data,bins=20)
plt.show()
```

```python
#descriptive plots
#bar plot
#default width is 0.8
values=[45,23,45,56,78,12]
vals=['MUM','DLH','HYD','BBS','MAA','PKD']
plt.figure(figsize=(10,4))
plt.bar(vals,values,width=0.5,color='rgbygb')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(8,8))
plt.pie(values,labels=vals)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(values,labels=vals,explode=[0,0,0.1,0,0,0],colors=['red','magenta','green','blue','lightblue','yellow'])
plt.show()
```
